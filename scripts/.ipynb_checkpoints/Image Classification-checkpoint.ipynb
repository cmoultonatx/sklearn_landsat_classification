```python
#Only new import is sklearn
import gdal
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Again, change the in_tif variable to the output path of the previous script
in_tif = r"..\..\\data\Austin_Landsat.tif"
in_ds = gdal.Open(in_tif)
arr_list = []
#Adds each band as an array
#I'm sure there's a more efficient way to get a GDAL raster as an n-dimensional array tho...
for i in range(1, in_ds.RasterCount +1):
    band = in_ds.GetRasterBand(i)
    arr = band.ReadAsArray()
    arr_list.append(arr)
in_data = np.dstack(np.array(arr_list)).T
```

```python
#Reshapes to a 2-D array
#TODO: Deal with the masked out values, the black outlines always count as 1 class
samples = in_data.reshape((in_data.shape[0],-1)).T
```

```python
#Sets up a KMeans classifier with 3 clusters (eg land surface classifications)
#This will take a little while to run...
kmeans = KMeans(n_clusters=4).fit_predict(samples)
```

```python
#Labels assign each pixel a class
labels= kmeans.reshape(in_data.shape[1], in_data.shape[2])
```

```python
#Let's take a look shall we :)
plt.figure(figsize=(20,20))
plt.imshow(np.stack(labels).T, cmap='Set3')
```
