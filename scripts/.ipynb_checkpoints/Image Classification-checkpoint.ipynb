```python
import gdal, ogr, osr
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
in_tif = r"C:\Users\moult\Documents\geospatial\data\Austin_Landsat.tif"
in_ds = gdal.Open(in_tif)
arr_list = []
for i in range(1, in_ds.RasterCount +1):
    band = in_ds.GetRasterBand(i)
    arr = band.ReadAsArray()
    arr_list.append(arr)
in_data = np.dstack(np.array(arr_list)).T
```

```python
samples = in_data.reshape((4,-1)).T
```

```python
kmeans = KMeans(n_clusters=4).fit_predict(samples)
```

```python
labels= kmeans.reshape(in_data.shape[1], in_data.shape[2])
```

```python
plt.figure(figsize=(20,20))
plt.imshow(np.stack(labels).T, cmap='Set3')
```
